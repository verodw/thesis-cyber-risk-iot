```python
import re

def extract_cpe_product(nodes):
    cpe_list = []

    def recurse(node):
        if isinstance(node, dict):
            if "cpe_match" in node:
                for cpe in node["cpe_match"]:
                    cpe_uri = cpe.get("cpe23Uri", "")
                    if cpe_uri:
                        parts = cpe_uri.split(":")
                        if len(parts) > 4:
                            cpe_list.append(parts[4].lower())
            if "children" in node:
                for child in node["children"]:
                    recurse(child)
        elif isinstance(node, list):
            for child in node:
                recurse(child)

    recurse(nodes)
    return list(set(cpe_list)) if cpe_list else None

df_nvd_clean = pd.DataFrame({
    'CVE_ID': df_nvd['cve.CVE_data_meta.ID'],
    'Description': df_nvd['cve.description.description_data'].apply(
        lambda x: x[0]['value'] if isinstance(x, list) and len(x) > 0 else None
    ),
    'CVSS_Score': df_nvd['impact.baseMetricV3.cvssV3.baseScore'],
    'Severity': df_nvd['impact.baseMetricV3.cvssV3.baseSeverity'],
    'Published': df_nvd['publishedDate'],
    'Last_Modified': df_nvd['lastModifiedDate'],
    'CPE_Product': df_nvd['configurations.nodes'].apply(extract_cpe_product) 
})
```

## Data Integration

```python
def perform_data_integration(df_shodan, df_nvd):
    """
    Integrasi data Shodan dan NVD berdasarkan produk yang sudah dinormalisasi dan di-clean.
    Menggunakan exact match, fuzzy match, substring, dan manual mapping.
    Prioritas match: exact > fuzzy > substring > manual
    """
    shodan_set = set(df_shodan['product_clean'].dropna().unique())
    nvd_set = set(df_nvd['product_clean'].dropna().unique())
    
    exact_matches = nvd_set.intersection(shodan_set)
    
    from difflib import SequenceMatcher
    def find_fuzzy_matches(nvd_set, shodan_set, threshold=0.8):
        matches = []
        for nvd_prod in nvd_set:
            if not nvd_prod or len(nvd_prod) < 3: 
                continue
            for shodan_prod in shodan_set:
                if not shodan_prod or len(shodan_prod) < 3: 
                    continue
                if SequenceMatcher(None, nvd_prod, shodan_prod).ratio() >= threshold:
                    matches.append((nvd_prod, shodan_prod))
        return matches
    fuzzy_matches = find_fuzzy_matches(nvd_set - exact_matches, shodan_set - exact_matches, threshold=0.7)
    
    def find_substring_matches(nvd_set, shodan_set, min_length=4):
        matches = []
        for nvd_prod in nvd_set:
            if not nvd_prod or len(nvd_prod) < min_length: 
                continue
            for shodan_prod in shodan_set:
                if not shodan_prod or len(shodan_prod) < min_length: 
                    continue
                if nvd_prod in shodan_prod or shodan_prod in nvd_prod:
                    matches.append((nvd_prod, shodan_prod))
        return matches
    substring_matches = find_substring_matches(nvd_set - exact_matches, shodan_set - exact_matches)
    
    manual_mapping = {
        'ssh': ['openssh','ssh'],
        'mysql': ['mysql','mariadb'],
        'http': ['apache','httpd','nginx','iis']
    }
    def find_manual_matches(nvd_set, shodan_set, mapping_dict):
        matches = []
        for _, variations in mapping_dict.items():
            nvd_matches = [p for p in nvd_set if any(var in p for var in variations)]
            shodan_matches = [p for p in shodan_set if any(var in p for var in variations)]
            for nvd_prod in nvd_matches:
                for shodan_prod in shodan_matches:
                    matches.append((nvd_prod, shodan_prod))
        return matches
    manual_matches = find_manual_matches(nvd_set - exact_matches, shodan_set - exact_matches, manual_mapping)
    
    mapping_final = {}
    match_dict = {}
    
    for prod in exact_matches:
        mapping_final[prod] = prod
        match_dict[prod] = 'exact'
    for nvd, shodan in fuzzy_matches:
        if nvd not in mapping_final:
            mapping_final[nvd] = shodan
            match_dict[nvd] = 'fuzzy'
    for nvd, shodan in substring_matches:
        if nvd not in mapping_final:
            mapping_final[nvd] = shodan
            match_dict[nvd] = 'substring'
    for nvd, shodan in manual_matches:
        if nvd not in mapping_final:
            mapping_final[nvd] = shodan
            match_dict[nvd] = 'manual'
    
    all_matches = set(mapping_final.keys())
    
    df_nvd_mapped = df_nvd.copy()
    df_nvd_mapped['product_mapped'] = df_nvd_mapped['product_clean'].map(mapping_final)
    
    df_integrated = pd.merge(
        df_shodan,
        df_nvd_mapped,
        left_on='product_clean',
        right_on='product_mapped',
        how='left',
        suffixes=('_shodan', '_nvd')
    )
    df_integrated['product_clean'] = df_integrated['product_clean_shodan']
    df_integrated['match_types'] = df_integrated['product_mapped'].map(match_dict)
    df_integrated['matched'] = df_integrated['CVE_ID'].notna()
    
    df_summary = df_integrated.groupby('product_clean').agg({
        'ip_address': 'nunique',
        'CVE_ID': 'nunique',
        'CVSS_Score_Final': 'max',
        'Severity_Final': lambda x: ','.join(set(x.dropna())),
        'matched': 'max'
    }).reset_index()
    
    df_summary.rename(columns={
        'ip_address': 'unique_ips',
        'CVE_ID': 'total_cves',
        'CVSS_Score_Final': 'max_cvss',
        'Severity_Final': 'severity_levels'
    }, inplace=True)
    
    print(f"Number of unique NVD products: {len(nvd_set)}")
    print(f"Number of unique Shodan products: {len(shodan_set)}")
    print(f"Total potential matches: {len(all_matches)}")
    print(f"Total integrated rows: {len(df_integrated)}")
    
    return df_integrated, df_summary
```

## Feature Engineering

```python
def feature_engineering(df_integrated, df_summary=None):
    import math
    import pandas as pd
    from sklearn.preprocessing import StandardScaler

    df = df_integrated[['device_id','ip_address','port','product_clean']].copy()

    max_cvss = df_integrated.groupby('device_id')['CVSS_Score_Final'].max().reset_index()
    max_cvss.rename(columns={'CVSS_Score_Final':'max_cvss_score'}, inplace=True)

    avg_cvss = df_integrated.groupby('device_id')['CVSS_Score_Final'].mean().reset_index()
    avg_cvss.rename(columns={'CVSS_Score_Final':'avg_cvss_score'}, inplace=True)

    total_vuln = df_integrated.groupby('device_id')['CVE_ID'].nunique().reset_index()
    total_vuln.rename(columns={'CVE_ID':'total_vulnerabilities'}, inplace=True)

    severity_dist = (
        df_integrated.groupby(['device_id','Severity_Final'])
        .size()
        .unstack(fill_value=0)
        .reset_index()
    )
    for sev in ['LOW','MEDIUM','HIGH','CRITICAL']:
        if sev not in severity_dist.columns:
            severity_dist[sev] = 0
    severity_dist.rename(columns={
        'LOW':'low_count','MEDIUM':'medium_count',
        'HIGH':'high_count','CRITICAL':'critical_count'
    }, inplace=True)

    open_ports = df_integrated.groupby('device_id')['port'].nunique().reset_index()
    open_ports.rename(columns={'port':'open_ports_count'}, inplace=True)

    df_features = (
        df[['device_id','ip_address','port','product_clean']]
        .drop_duplicates()
        .merge(max_cvss, on='device_id', how='left')
        .merge(avg_cvss, on='device_id', how='left')
        .merge(total_vuln, on='device_id', how='left')
        .merge(severity_dist, on='device_id', how='left')
        .merge(open_ports, on='device_id', how='left')
        .fillna(0)
    )
    df_features['risk_index'] = df_features.apply(
        lambda row: row['avg_cvss_score'] * math.log(row['total_vulnerabilities'] + 1),
        axis=1
    )
    df_features['critical_flag'] = (df_features['max_cvss_score'] >= 9).astype(int)
    df_features = pd.get_dummies(df_features, columns=['product_clean'], prefix='dtype')

    numeric_cols = [
        'port','total_vulnerabilities','avg_cvss_score','max_cvss_score',
        'risk_index','critical_flag','open_ports_count',
        'low_count','medium_count','high_count','critical_count'
    ]

    scaler = StandardScaler()
    df_features[numeric_cols] = scaler.fit_transform(df_features[numeric_cols])

    return df_features
```

```python
def run_pipeline(df_shodan, df_nvd_clean):
    df_integrated, df_summary = perform_data_integration(df_shodan, df_nvd_clean)
    print(f"\n=== Data Integration ===")
    print(f"Number of devices after integration    : {df_integrated['device_id'].nunique()}")
    print(f"Number of integrated rows              : {len(df_integrated)}")
    print(f"Number of unique Shodan products       : {df_shodan['product_clean'].nunique()}")
    print(f"Number of unique NVD products          : {df_nvd_clean['product_clean'].nunique()}")
    print("\nExample of Integrated Data:")
    print(df_integrated.head())

    df_features = feature_engineering(df_integrated, df_summary)
    print(f"\n=== Feature Engineering ===")
    print(f"Number of devices after FE      : {df_features['device_id'].nunique()}")
    print(f"Number of feature rows          : {len(df_features)}")
    print(f"Number of feature columns       : {df_features.shape[1]}")
    print("\nColumn Name Features:")
    print(df_features.columns.tolist())
    print("\nExample of Engineered Features:")
    print(df_features.head())

    return df_integrated, df_summary, df_features

df_integrated, df_summary, df_features = run_pipeline(df_shodan, df_nvd_clean)
```

## Feature Selection & Split Dataset

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

def feature_selection_for_vae(df_features, variance_thresh=0.0, keep_cols=None):
    """
    Feature selection untuk VAE: pertahankan cukup fitur numerik.
    - Hanya hapus fitur dengan variance sangat rendah.
    - keep_cols wajib dipertahankan.
    """
    if keep_cols is None:
        keep_cols = []

    numeric_cols = [
        'port', 'total_vulnerabilities', 'avg_cvss_score', 'max_cvss_score',
        'risk_index','critical_flag','open_ports_count',
        'low_count','medium_count','high_count','critical_count'
    ]
    numeric_cols = [c for c in numeric_cols if c in df_features.columns]
    binary_cols = [c for c in df_features.columns if c.startswith('dtype_')]

    df_numeric = df_features[numeric_cols].fillna(0)
    vt = VarianceThreshold(threshold=variance_thresh)
    vt.fit(df_numeric)
    selected_numeric = df_numeric.columns[vt.get_support()].tolist()

    final_numeric = [f for f in selected_numeric if f not in keep_cols]
    final_features = final_numeric + binary_cols

    id_cols = [c for c in ['device_id', 'ip_address'] if c in df_features.columns]
    df_selected = df_features[id_cols + final_features + keep_cols].copy()

    summary_dict = {
        "Number of Devices": [df_features['device_id'].nunique(), df_selected['device_id'].nunique()],
        "Number of Rows": [len(df_features), len(df_selected)],
        "Number of Columns": [df_features.shape[1], df_selected.shape[1]],
        "Numeric Features": [len(selected_numeric), len(final_numeric)],
        "Total Features (numeric + binary + target)": [len(selected_numeric) + len(binary_cols), len(final_features) + len(keep_cols)]
    }
    print("\n=== FEATURE SELECTION ===")
    print(df_summary)

    print("\n[FEATURE SELECTION FOR VAE]")
    print(f"Number of numerical features used (without target): {len(final_numeric)}")
    print(f"Example of numerical features: {final_numeric[:10]}")
    print(f"Total number of features (numeric + binary + target): {len(final_features + keep_cols)}")

    return df_selected, final_numeric, df_summary

def split_dataset_for_vae(df_selected, numeric_features, vuln_col='total_vulnerabilities',
                          vuln_threshold=2, test_size=0.2, random_state=42):
    """
    Split dataset untuk VAE deep learning:
    - Train: hanya normal (total_vulnerabilities <= threshold)
    - Test: normal + anomaly
    - Hanya fitur numerik dipakai
    - Include StandardScaler
    """
    if vuln_col not in df_selected.columns:
        raise KeyError(f"Kolom '{vuln_col}' tidak ada di df_selected.")

    X = df_selected[numeric_features]  

    normal_mask = df_selected[vuln_col] <= vuln_threshold
    anomaly_mask = df_selected[vuln_col] > vuln_threshold

    X_normal = X[normal_mask].copy()
    X_anomaly = X[anomaly_mask].copy()
    X_train, X_test_norm = train_test_split(
        X_normal, test_size=test_size, random_state=random_state
    )
    X_test = pd.concat([X_test_norm, X_anomaly], axis=0).reset_index(drop=True)
    y_test = np.concatenate([
        np.zeros(len(X_test_norm)),   # normal = 0
        np.ones(len(X_anomaly))       # anomaly = 1
    ])
    y_test = pd.Series(y_test, name='label')

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    print("\n[DATA SPLIT FOR VAE]")
    print(f"Train shape (normal only): {X_train.shape}")
    print(f"Test shape: {X_test.shape}")
    print(f"Normal in test: {(y_test==0).sum()}, Anomaly in test: {(y_test==1).sum()}")
    print(f"Numerical features for VAE: {numeric_features}")

    return X_train_scaled, X_test_scaled, y_test, scaler

def run_pipeline_for_vae(df_shodan, df_nvd_clean, vuln_threshold=2):
    """
    Pipeline lengkap sampai split dataset untuk VAE.
    """
    df_integrated, df_summary_int = perform_data_integration(df_shodan, df_nvd_clean)
    df_features = feature_engineering(df_integrated, df_summary_int)

    df_selected, numeric_features, df_summary_fs = feature_selection_for_vae(
        df_features, keep_cols=['total_vulnerabilities']
    )

    X_train, X_test, y_test, scaler = split_dataset_for_vae(
        df_selected, numeric_features, vuln_col='total_vulnerabilities', vuln_threshold=vuln_threshold
    )

    return df_integrated, df_summary_int, df_features, df_selected, numeric_features, X_train, X_test, y_test, scaler, df_summary_fs

df_integrated, df_summary_int, df_features, df_selected, numeric_features, X_train, X_test, y_test, scaler, df_summary_fs = run_pipeline_for_vae(df_shodan, df_nvd_clean)
```

## Apply VAE Model 

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, precision_recall_fscore_support, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

class SimpleVAE(tf.keras.Model):
    def __init__(self, input_dim, latent_dim=2, beta=0.005):
        super(SimpleVAE, self).__init__()
        self.latent_dim = latent_dim
        self.beta = beta
        
        self.encoder_h1 = layers.Dense(max(8, input_dim*2), activation='relu')
        self.encoder_h2 = layers.Dense(max(4, input_dim), activation='relu')
        self.z_mean = layers.Dense(latent_dim)
        self.z_log_var = layers.Dense(latent_dim)
        
        self.decoder_h1 = layers.Dense(max(4, input_dim), activation='relu')
        self.decoder_h2 = layers.Dense(max(8, input_dim*2), activation='relu')
        self.decoder_output = layers.Dense(input_dim, activation='linear')
        
    def encode(self, x):
        h = self.encoder_h1(x)
        h = self.encoder_h2(h)
        return self.z_mean(h), self.z_log_var(h)
    
    def reparameterize(self, z_mean, z_log_var):
        epsilon = tf.random.normal(shape=tf.shape(z_mean))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
    
    def decode(self, z):
        h = self.decoder_h1(z)
        h = self.decoder_h2(h)
        return self.decoder_output(h)
    
    def call(self, inputs):
        z_mean, z_log_var = self.encode(inputs)
        z = self.reparameterize(z_mean, z_log_var)
        reconstruction = self.decode(z)
        
        kl_loss = -0.5 * tf.reduce_mean(
            1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
        )
        self.add_loss(self.beta * kl_loss)
        return reconstruction

def train_and_evaluate_vae(X_train, X_test, y_test, epochs=50, batch_size=16, latent_dim=2, beta=0.005):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    vae = SimpleVAE(input_dim=X_train.shape[1], latent_dim=latent_dim, beta=beta)
    vae.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
    lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(monitor='loss', factor=0.5, patience=5, min_lr=1e-5, verbose=1)
    
    history = vae.fit(
        X_train_scaled, X_train_scaled,
        epochs=epochs,
        batch_size=batch_size,
        shuffle=True,
        verbose=1,
        callbacks=[lr_scheduler]
    )
    
    X_test_recon = vae.predict(X_test_scaled, verbose=0)
    recon_errors = np.mean((X_test_scaled - X_test_recon)**2, axis=1)
    X_train_recon = vae.predict(X_train_scaled, verbose=0)
    train_recon_errors = np.mean((X_train_scaled - X_train_recon)**2, axis=1)
    
    threshold = np.mean(train_recon_errors) + 2 * np.std(train_recon_errors)
    y_pred = (recon_errors > threshold).astype(int)
    
    auc_score = roc_auc_score(y_test, recon_errors)
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary', zero_division=0)
    
    print("\n=== VAE Performance ===")
    print(f"F1 Score: {f1:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}, AUC: {auc_score:.3f}")
    
    return vae, scaler, threshold, history, recon_errors, y_pred

vae_model, scaler, recon_threshold, history, recon_errors, y_pred = train_and_evaluate_vae(
    X_train, X_test, y_test, epochs=50, batch_size=16, latent_dim=2, beta=0.005
)
```

```python
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score, classification_report, confusion_matrix, roc_curve, auc
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_recon = vae_model.predict(X_train_scaled, verbose=0)
train_recon_errors = np.mean((X_train_scaled - X_train_recon) ** 2, axis=1)

X_test_pred = vae_model.predict(X_test_scaled, verbose=0)
recon_errors = np.mean((X_test_scaled - X_test_pred) ** 2, axis=1)

train_recon_errors = np.log1p(train_recon_errors)
recon_errors = np.log1p(recon_errors)

fpr, tpr, thresholds = roc_curve(y_test, recon_errors)
j_scores = tpr - fpr
j_best = np.argmax(j_scores)
optimal_threshold = thresholds[j_best]

print(f"Optimal Threshold (Youden J): {optimal_threshold:.6f}")

y_pred = (recon_errors > optimal_threshold).astype(int)

def validate_vae_simple(y_test, y_pred, recon_errors, threshold, fpr, tpr):
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="binary")
    roc_auc = roc_auc_score(y_test, recon_errors)

    metrics_df = pd.DataFrame({
        "Metric": ["Precision", "Recall", "F1-score", "ROC-AUC"],
        "Value": [precision, recall, f1, roc_auc]
    })
    print("\n=== Performance Summary ===")
    print(metrics_df)

    report = classification_report(y_test, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    print("\n=== Classification Report ===")
    print(report_df.round(3))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Normal","Anomaly"], yticklabels=["Normal","Anomaly"])
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

    plt.figure(figsize=(6,4))
    sns.boxplot(x=y_test, y=recon_errors, palette="Set2")
    plt.axhline(y=threshold, color="red", linestyle="--", label="Threshold")
    plt.xticks([0,1], ["Normal","Anomaly"])
    plt.ylabel("Reconstruction Error")
    plt.title("Reconstruction Error by Class")
    plt.legend()
    plt.show()

    plt.figure(figsize=(5,4))
    sns.countplot(x=y_pred, palette="coolwarm")
    plt.xticks([0,1], ["Normal","Anomaly"])
    plt.title("Predicted Class Distribution")
    plt.show()

    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"ROC curve (AUC = {auc(fpr,tpr):.3f})")
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

validate_vae_simple(y_test, y_pred, recon_errors, optimal_threshold, fpr, tpr)
```

## Risk Scoring

```python
def calculate_risk_scores_clean(df_results, recon_errors, cve_data=None):
    """
    Hitung risk score unik per device:
    - recon_error dinormalisasi
    - cvss_score diambil max per product (biar gak duplikat)
    - risk_score gabungan recon_norm + CVSS severity
    """

    df_results = df_results.copy()
    df_results["recon_error"] = recon_errors
    df_results["recon_norm"] = (
        (df_results["recon_error"] - df_results["recon_error"].min()) /
        (df_results["recon_error"].max() - df_results["recon_error"].min())
    )

    if cve_data is not None and "product_normalized" in df_results.columns and "product_normalized" in cve_data.columns:
        cve_agg = cve_data.groupby("product_normalized")["cvss_score"].max().reset_index()
        df_risk = df_results.merge(cve_agg, on="product_normalized", how="left")
        df_risk["cvss_score"] = df_risk["cvss_score"].fillna(0)
    else:
        df_risk = df_results.copy()
        df_risk["cvss_score"] = 0

    df_risk["risk_score"] = 0.6 * df_risk["recon_norm"] + 0.4 * (df_risk["cvss_score"] / 10)
    df_final = df_risk.groupby("device_id", as_index=False).agg({
        "y_true": "max",
        "y_pred": "max",
        "recon_error": "mean",
        "recon_norm": "mean",
        "cvss_score": "max",
        "risk_score": "max"
    })

    return df_final
```

## CVE Matching

```python
from rapidfuzz import process, fuzz

def perform_cve_matching(df_risk, df_shodan_prepared, df_nvd_clean, threshold=85):
    choices = df_nvd_clean['product_clean'].dropna().unique()
    mapping_dict = {
        "mikrotik": "routeros",
        "nginx": "nginx",
        "mysql": "mysql",
        "apache": "apache_http_server",
        "openssh": "openssh",
        "mosquitto": "eclipse_mosquitto",
        "hikvision": "hikvision",
        "dahua": "dahua",
        "netgear": "netgear",
        "tenda": "tenda",  
    }

    def get_best_match(product):
        if not isinstance(product, str) or product.strip() == "":
            return None
        p = product.lower().strip()

        if p in mapping_dict:
            return mapping_dict[p]

        match = process.extractOne(p, choices, scorer=fuzz.token_sort_ratio)
        if match:
            candidate, score, _ = match
            if score >= threshold and len(candidate) >= 3:
                return candidate
        return None

    df_shodan_prepared = df_shodan_prepared.copy()
    df_shodan_prepared["matched_product"] = df_shodan_prepared["product_normalized"].apply(get_best_match)

    df_device_risk = df_risk.merge(
        df_shodan_prepared[['device_id','matched_product','organization','city']],
        on='device_id',
        how='left'
    )
    df_nvd_clean['severity_levels'] = pd.cut(
        df_nvd_clean['CVSS_Score_Final'],
        bins=[0, 4, 7, 9.9, 10],
        labels=['LOW','MEDIUM','HIGH','CRITICAL'],
        right=False
    )
    df_matched = df_device_risk.merge(
        df_nvd_clean[['product_clean','CVE_ID','CVSS_Score_Final','severity_levels','Description']],
        left_on='matched_product',
        right_on='product_clean',
        how='left'
    )
    df_matched['product_clean'] = df_matched['product_clean'].fillna(df_matched['matched_product'])
    df_matched['severity_levels'] = df_matched['severity_levels'].fillna('LOW')
    df_matched['CVSS_Score_Final'] = df_matched['CVSS_Score_Final'].fillna(0)
    df_matched['CVE_ID'] = df_matched['CVE_ID'].fillna('None')
    df_matched['Description'] = df_matched['Description'].fillna('No CVE found')

    return df_matched

df_cve_matched = perform_cve_matching(df_risk, df_shodan_prepared, df_nvd_clean)
print(df_cve_matched.head(20))
```

## Risk Profile Creation

```python
import pandas as pd

def create_risk_profile(df_cve_matched, include_timestamp=True):
    """
    Membuat Risk Profile per device.
    df_cve_matched: hasil dari perform_cve_matching
    include_timestamp: True jika ingin menambahkan kolom last_seen
    """
    df_profile = df_cve_matched.groupby(
        ['device_id','organization','city','matched_product'], as_index=False
    ).agg({
        'CVSS_Score_Final':'max', 
        'severity_levels':'max', 
        'CVE_ID': lambda x: ', '.join(x.dropna().unique()),  
        'risk_score':'first',       
        'label_str':'first',        
    })
    df_profile = df_profile.rename(columns={
        'severity_levels':'risk_level',
        'CVE_ID':'CVE_References'
    })
    if include_timestamp and 'last_seen' in df_cve_matched.columns:
        df_profile = df_profile.merge(
            df_cve_matched[['device_id','last_seen']].drop_duplicates(),
            on='device_id',
            how='left'
        )
    cols_order = ['device_id','organization','city','matched_product','risk_score','label_str','risk_level','CVE_References']
    if include_timestamp and 'last_seen' in df_profile.columns:
        cols_order.append('last_seen')

    df_profile = df_profile[cols_order]

    return df_profile

df_risk_profile = create_risk_profile(df_cve_matched)
df_risk_profile.head(20)
```
